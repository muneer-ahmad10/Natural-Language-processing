```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = [
    "Thor eating pizza, Loki is eating pizza, Ironman ate pizza already",
    "Apple is announcing new iphone tomorrow",
    "Tesla is announcing new model-3 tomorrow",
    "Google is announcing new pixel-6 tomorrow",
    "Microsoft is announcing new surface tomorrow",
    "Amazon is announcing new eco-dot tomorrow",
    "I am eating biryani and you are eating grapes"
]
```

```python
v=TfidfVectorizer()
trans_output=v.fit_transform(corpus)
print(v.vocabulary_)
```

```python
print(dir(v))
```

```python
all_feature_name=v.get_feature_names_out()

for word in all_feature_name:
  idx=v.vocabulary_.get(word)
  idf_score = v.idf_[idx]
  print(f'{word} : {idf_score}')
```

```python
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

from sklearn.naive_bayes import MultinomialNB

from sklearn.ensemble import RandomForestClassifier
```

```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/NLP/DataSets/ecommerceDataset.csv",names=["Category","Text"])
```

```python
df.head(5)
```

```python
df['Category_num']=df['Category'].map({
   'Household' : 0,
   'Books' : 1,
   'Electronics' : 2,
   'Clothing & Accessories' : 3

})
```

```python
df['Category'].value_counts()
```

balancing Dataset using undersampling

```python
df_household=df[df['Category'] == 'Household'].sample(8000)
df_Books=df[df['Category'] == 'Books'].sample(8000)
df_Electronics=df[df['Category'] == 'Electronics'].sample(8000)
df_Accessories=df[df['Category'] == 'Clothing & Accessories'].sample(8000)
```

```python
df=pd.concat([df_household,df_Books,df_Electronics,df_Accessories],axis=0)
df = df.reset_index(drop=True)
```

```python
df=df.dropna()
```

```python
X=df['Text']
y=df['Category_num']
```

# **Train_test_split**

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=2000,stratify=df.Category_num)
```

```python
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

```python
y_train.value_counts()
```

### **using sklearn pipeline module create a classification pipeline to classify the Ecommerce Data.**

use TF-IDF for pre-processing the text.

use KNN as the classifier

print the classification report.

```python



clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),
     ('KNN', KNeighborsClassifier())
])

clf.fit(X_train, y_train)


y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```

# **Using text pre-processing to remove stop words, punctuations and apply lemmatization**

```python

import spacy

nlp = spacy.load("en_core_web_sm")

def preprocess(text):

    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)

    return " ".join(filtered_tokens)
```

```python
df['preprocessed_txt'] = df['Text'].apply(preprocess)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.preprocessed_txt,
    df.Category_num	,
    test_size=0.2,
    random_state=2022,
    stratify=df.Category_num
)
```

```python
clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),
     ('Random Forest', RandomForestClassifier())
])


clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)



print(classification_report(y_test, y_pred))
```
