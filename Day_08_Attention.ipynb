# **Scaled Dot Product Attention**

```python
# when the vectors become large dot product becomes huge e.g vectors are 100,200,500 softmax becomes extreme for 500 softmax will be 0.999999 and dor 100 it will be 0.00001 this means model become more confident and training becomes unstable


# to solve this problem we scale scores down and here transformer paper introduced. Attention
# (𝑄,𝐾,𝑉)=softmax(𝑄𝐾.𝑇 / sqrt(𝑑𝑘))𝑉

# dk the dimension of key vectors which is taken out using K.shape[-1] we perform sq.root to it and divide it to the scores .scores=𝑄𝐾.𝑇
```

```python
embeddings = torch.tensor([
    [1.0, 0.0, 1.0, 0.0],  # I
    [0.0, 2.0, 0.0, 2.0],  # love
    [1.0, 1.0, 1.0, 1.0]   # NLP
])

# Q K V
Q = embeddings
K = embeddings
V = embeddings

# Similarity scores
scores = torch.matmul(Q, K.T)

#Scaling
dk=K.shape[-1]
scaled_scores=scores / torch.sqrt(torch.tensor(dk,dtype=torch.float32))

# Attention weights

attention_weights = f.softmax(scaled_scores, dim=1)

# Final output
output = torch.matmul(attention_weights, V)

output = torch.matmul(attention_weights, V)

print("Scores:\n", scores)

print("\nScaled Scores:\n", scaled_scores)

print("\nAttention Weights:\n", attention_weights)

print("\nFinal Output:\n", output)
```

```python
# if embedding dimension increases scores become very large scaling prevents softmax saturation
```
