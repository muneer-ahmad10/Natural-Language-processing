Example

Document:

Chunk 1:
GPT-4o-mini was released by OpenAI.

Chunk 2:
Large language models are useful for AI applications.

Question:

What is GPT-4o-mini?

Sometimes FAISS may think:

Chunk 2

is semantically similar because of "AI" and "language models".

But the user actually needs:

Chunk 1

because it contains the exact keyword.

**BM25**

BM25 is a keyword-based retrieval algorithm.

Instead of vectors, it checks:

How many query words appear in the document?
How rare are those words?

For:

GPT-4o-mini

BM25 loves exact matches.

It will strongly favor:

GPT-4o-mini was released by OpenAI.

Hybrid Search

Instead of choosing one:

FAISS only ❌
BM25 only ❌

Use both:

BM25 + FAISS ✅

Pipeline:

Question
    ↓
BM25 Results
    ↓
FAISS Results
    ↓
Merge Results
    ↓
CrossEncoder
    ↓
LLM

This gives:

Exact keyword matching (BM25)
Semantic matching (FAISS)
Accurate ranking (CrossEncoder)
