Example

Document:

Chunk 1:
GPT-4o-mini was released by OpenAI.

Chunk 2:
Large language models are useful for AI applications.

Question:

What is GPT-4o-mini?

Sometimes FAISS may think:

Chunk 2

is semantically similar because of "AI" and "language models".

But the user actually needs:

Chunk 1

because it contains the exact keyword.

**BM25**

BM25 is a keyword-based retrieval algorithm.

Instead of vectors, it checks:

How many query words appear in the document?
How rare are those words?

For:

GPT-4o-mini

BM25 loves exact matches.

It will strongly favor:

GPT-4o-mini was released by OpenAI.

Hybrid Search

Instead of choosing one:

FAISS only ❌
BM25 only ❌

Use both:

BM25 + FAISS ✅

Pipeline:

Question
    ↓
BM25 Results
    ↓
FAISS Results
    ↓
Merge Results
    ↓
CrossEncoder
    ↓
LLM

This gives:

Exact keyword matching (BM25)
Semantic matching (FAISS)
Accurate ranking (CrossEncoder)

```python
from rank_bm25 import BM25Okapi
```

```python
documents = [
    "GPT-4o-mini was released by OpenAI",
    "Python is a programming language",
    "Football is a popular sport"
]
```

```python
tokenized_doc=[
    token.split()
    for token in documents
]
```

```python
tokenized_doc
```

```python
bm25 = BM25Okapi(tokenized_doc)
```

```python
query = "GPT-4o-mini"
```

```python
tokenized_query=query.split()
```

```python
tokenized_query
```

```python
scores = bm25.get_scores(
    tokenized_query
)

print(scores)
```

```python
import numpy as np


highest_score_idx = np.argmax(scores)


highest_score = scores[highest_score_idx]
winning_chunk = documents[highest_score_idx]

print(f"Highest Score: {highest_score}")
print(f"Top Document: '{winning_chunk}'")
```
