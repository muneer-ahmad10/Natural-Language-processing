# **Sentiment**


```python
from transformers import pipeline
clf = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")
result = clf("I absolutely loved this film!")
print(result)
```

## **PipeLine**
tokenize → run model → take argmax of output logits → map label index to string → return list of dicts with label and score

# **NER**

NER stands for Named Entity Recognition. It is a crucial task in Natural Language Processing (NLP) that acts like a "highlighter" for text.

Its job is to automatically read a sentence, locate real-world objects (entities), and classify them into predefined categories.

```python
ner = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", aggregation_strategy="simple")
result = ner("Elon Musk founded SpaceX in California.")
print(result)
```

## Pipeline
tokenize → run token classification model → each token gets a label (B-PER, I-PER, O…) → aggregation_strategy merges subword tokens into whole entities

# **QA**

```python
qa = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
result = qa( question="What does BERT stand for?", context="BERT stands for Bidirectional Encoder Representations from Transformers." )
print(result)
```

## **Pipeline**
tokenize [CLS] question [SEP] context [SEP] → model predicts start and end token positions → extract span from context → return answer string + confidence score
