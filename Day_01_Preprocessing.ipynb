# **Raw Data**

## **Sample text**

```python
text='I am loving the new AI tchnologies'
```

# **Preprocessing**

## **Tokenization**

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(text)
print(tokens)
```

## **Lower case**

```python
tokens=[word.lower() for word in tokens]
print(tokens)
```

## **stop word removal**

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words('english'))
filtered=[words for words in tokens if words not in stop_words]
print(filtered)
```

## **Stemming**

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
stemmed=[stemmer.stem(word) for word in filtered]
print(stemmed)
```



*   in this stemmed function technologies became tchnolog (not real word)
*   to avoid this we use lemmetization



## **Lemmetization**

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmetizer=WordNetLemmatizer()
lemetized=[lemmetizer.lemmatize(word) for word in filtered]
print(lemetized)
```

# Full pipeline of preprocessing

```python
def preprocess(text):

  from nltk.tokenize import word_tokenize
  from nltk.corpus import stopwords
  from nltk.stem import WordNetLemmatizer

  tokens= word_tokenize(text.lower())

  stop_words=set(stopwords.words('english'))

  tokens=[w for w in tokens if w.isalpha()]
  filtered=[word for word in tokens if word not in stop_words]

  lemmetizer = WordNetLemmatizer()
  tokens=[lemmetizer.lemmatize(w) for w in filtered]

  return tokens
```

```python
text=preprocess('hello my name is abrar i am from jammu and kashmir i am doing BCA')
```
