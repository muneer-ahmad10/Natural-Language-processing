**Fine Tuning :** means taking pretrained model and training it further on a smallar ,task specific dataset

in simple words:
**pre_trained model** + **specific Training Data** = **Fine Tuned Model**

eg:
A general language model may understand english well
but  if we tune it on customer support emails it can become better at understanding support related questions

Why do we fine tune model ?

fine tune is usefull when we want a model to perform better for a specific use case

Examples:

Use Case	Fine-Tuning Goal

Customer support	Classify tickets correctly

Healthcare	Understand medical text better

Education	Evaluate student responses

Banking	Detect fraud-related messages

E-commerce	Understand product reviews

Fine-tuning helps the model become more focused and accurate for the required task.

A pre-trained model may understand general text.

A fine-tuned model can understand company-specific support tickets, policies, or product-related language better.



SImple fine tuning workflow :

Collect task-specific data

        ↓

Clean and prepare the data

        ↓

Load a pre-trained transformer model

        ↓

Train the model on the new dataset

        ↓

Evaluate the model

        ↓
        
Use it for predictions

```python
from datasets import Dataset
data = {
    "question": [
        "What is Python?",
        "What is Machine Learning?"
    ],
    "answer": [
        "Python is a programming language.",
        "Machine Learning enables computers to learn from data."
    ]
}

dataset=Dataset.from_dict(data)
print(dataset)
```

```python
from transformers import(
    AutoTokenizer,
    AutoModelForSeq2SeqLM
)
```

```python
model_name='google/Flan-t5-base'

tokenizer=AutoTokenizer.from_pretrained(model_name)

model=AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
def preprocess(example):
  input_text=example['question']
  target_text=example['answer']


  model_inputs=tokenizer(
      input_text,
      max_length=128,
      truncation=True
  )

  labels=tokenizer(
      target_text,
      max_length=128,
      truncation=True

  )

  model_inputs['labels']=labels['input_ids']

  return model_inputs
```

```python
tokenized_dataset=dataset.map(
    preprocess
)
```

```python
tokenized_dataset
```

```python
question = "What is Python?"

inputs = tokenizer(
    question,
    return_tensors="pt"
)

outputs = model.generate(
    **inputs
)

print(
    tokenizer.decode(
        outputs[0],
        skip_special_tokens=True
    )
)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=20,
    per_device_train_batch_size=2,
    logging_steps=1,
    save_strategy="no"
)
```

```python
from transformers import Trainer

from transformers import DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
)

```

```python
trainer.train()
```
