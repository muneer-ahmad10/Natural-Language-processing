**Fine Tuning :** means taking pretrained model and training it further on a smallar ,task specific dataset

in simple words:
**pre_trained model** + **specific Training Data** = **Fine Tuned Model**

eg:
A general language model may understand english well
but  if we tune it on customer support emails it can become better at understanding support related questions

Why do we fine tune model ?

fine tune is usefull when we want a model to perform better for a specific use case

Examples:

Use Case	Fine-Tuning Goal

Customer support	Classify tickets correctly

Healthcare	Understand medical text better

Education	Evaluate student responses

Banking	Detect fraud-related messages

E-commerce	Understand product reviews

Fine-tuning helps the model become more focused and accurate for the required task.

A pre-trained model may understand general text.

A fine-tuned model can understand company-specific support tickets, policies, or product-related language better.
