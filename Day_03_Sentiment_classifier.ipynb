```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
# def preprocesses(text):

#   preprocessed_text=[]

#   lemmatizer=WordNetLemmatizer()
#   stop_words=set(stopwords.words('english'))

#   tokens=word_tokenize(text)
#   tokens=[w.lower() for w in tokens]
#   tokens=[w for w in tokens if w not in stop_words and w.isalpha()]
#   tokens=[lemmetizer.lemmatize(w) for w in tokens]

#   preprocessed_text.append(tokens)

#   return preprocessed_text


from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

def preprocesses(texts):  # now expects LIST
    preprocessed_text = []

    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))

    for text in texts:  # 🔥 loop over each sentence
        tokens = word_tokenize(text)
        tokens = [w.lower() for w in tokens]
        tokens = [w for w in tokens if w not in stop_words and w.isalpha()]
        tokens = [lemmatizer.lemmatize(w) for w in tokens]  # fixed typo

        preprocessed_text.append(tokens)

    return preprocessed_text



```

```python
texts = [
    "I love this movie",
    "This is amazing",
    "I hate this",
    "This is terrible",
    "I really like this",
    "I dislike this product"
]
labels = [1, 1, 0, 0, 1, 0]
```

```python
clean_text=preprocesses(texts)
clean_sentences = [" ".join(tokens) for tokens in clean_text]
```

```python
clean_sentences
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vect=TfidfVectorizer()
X=vect.fit_transform(clean_sentences)
```

```python
print(vect.get_feature_names_out())
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X,labels)
```

```python
test = ["I love this so much"]
test_vec = vect.transform(test)
pred=model.predict(test_vec)

print('Positive' if pred == 1 else 'Negative')
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
def preprocesses(text):
  processed=[]
  for sentence in text:
    doc=nlp(sentence)

    tokens=[token.lemma_.lower() for token in doc if not token.is_stop and token.is_alpha]
    processed.append(" ".join(tokens))

  return processed
```

```python
# texts = [
#     "I love this movie",
#     "This is amazing",
#     "I hate this",
#     "This is terrible",
#     "I really like this",
#     "I dislike this product"
# ]

# clean_texts = preprocesses(text)
# print(clean_texts)
texts = [
    "I love this movie",
    "This is amazing",
    "I hate this",
    "This is terrible",
    "I really like this",
    "I dislike this product"
]

clean_texts = preprocesses(texts)
print(clean_texts)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Vectorization
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(clean_texts)

# Labels
labels = [1, 1, 0, 0, 1, 0]

# Model
model = LogisticRegression()
model.fit(X, labels)
```

```python
test = ["I really love this"]

test_clean = preprocesses(test)
test_vec = vectorizer.transform(test_clean)
spacy_pred=model.predict(test_vec)
print(model.predict(test_vec))

print('Positive' if spacy_pred == 1 else 'Negative')

```
