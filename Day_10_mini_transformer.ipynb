```python
import torch
import torch.nn as nn
import math
```

```python
# ===================================
# INPUT TOKENS
# ===================================

tokens = torch.tensor([1,2,3])
```

```python
# ===================================
# EMBEDDING
# ===================================

embedding = nn.Embedding(
    num_embeddings=10,
    embedding_dim=4
)

x = embedding(tokens)
```

```python
# ===================================
# POSITIONAL ENCODING
# ===================================

position = torch.arange(3).unsqueeze(1)

x = x + position
```

```python
# ===================================
# Q K V
# ===================================

embed_dim = 4

Wq = nn.Linear(embed_dim, embed_dim)
Wk = nn.Linear(embed_dim, embed_dim)
Wv = nn.Linear(embed_dim, embed_dim)

Q = Wq(x)
K = Wk(x)
V = Wv(x)
```

```python
# ===================================
# ATTENTION SCORES
# ===================================

scores = torch.matmul(Q, K.T)

scaled_scores = scores / math.sqrt(embed_dim)

attention_weights = torch.softmax(
    scaled_scores,
    dim=-1
)
```

```python
# ===================================
# CONTEXT
# ===================================

context = torch.matmul(
    attention_weights,
    V
)
```

```python
# ===================================
# FEED FORWARD
# ===================================

ff = nn.Sequential(
    nn.Linear(embed_dim, 16),
    nn.ReLU(),
    nn.Linear(16, embed_dim)
)

output = ff(context)

print(output)
```
