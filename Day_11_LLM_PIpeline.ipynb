# 🚀 STEP 2 — TOKENIZATION

Convert text → tokens.

# 🚀 STEP 3 — EMBEDDINGS

Tokens become vectors.

token_id → embedding vector
🚀 EXAMPLE
102 → [0.2, 0.7, -0.1]

Now transformer can process meaning mathematically.
