# **NLTK**

```python
texts = [
    "I love this movie",
    "This is amazing",
    "I hate this",
    "This is terrible",
    "I really like this",
    "I dislike this product"
]

labels = [1, 1, 0, 0, 1, 0]
```

```python
import nltk
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression
```

## For **TFIDF**

```python
def preprocess(texts):
  stop_words=set(stopwords.words('english'))
  lemmatizer=WordNetLemmatizer()

  cleaned=[]

  for sentence in texts:
    tokens=word_tokenize(sentence.lower())
    tokens=[w for w in tokens if w.isalpha()]
    tokens=[w for w in tokens if w not in stop_words]
    tokens=[lemmatizer.lemmatize(w) for w in tokens]

    cleaned.append(' '.join(tokens))

  return cleaned
```

```python
clean_text=preprocess(texts)
print(clean_text)
```

```python
vect=TfidfVectorizer()
X=vect.fit_transform(clean_text)
```

```python
model = LogisticRegression()
model.fit(X, labels)
```

```python
test = ["I really love this"]

test_clean = preprocess(test)
test_vec = vect.transform(test_clean)

print(model.predict(test_vec))
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
def preprocess_spacy(texts):
  cleaned=[]

  for sent in texts :
    doc=nlp(sent)

    tokens=[
        token.lemma_.lower()
        for token in doc
        if not token.is_stop and token.is_alpha
    ]

    cleaned.append(' '.join(tokens))

  return cleaned
```

```python
spacy_clean=preprocess_spacy(texts)
print(spacy_clean)
```

```python
spacy_vect = TfidfVectorizer()
spacy_X=spacy_vect.fit_transform(spacy_clean)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(spacy_X, labels)
```

```python
test = ["I really hate this"]

test_clean = preprocess_spacy(test)
test_vec =  spacy_vect.transform(test_clean)

print(model.predict(test_vec))
```
