# **What is a Vector Database?**

A vector database stores:

Vector
+
Original Text
+
Metadata

Example:


```python
{
    "vector":[0.12,-0.55,...],
    "text":"GPT is a decoder-only model",
    "metadata":{
        "source":"transformers.pdf",
        "page":5
    }
}
```

# **Why not use SQL?**



SQL databases search:

Exact values

Example:

**SELECT * FROM documents
WHERE title = "GPT"**

But semantic search needs:

Meaning similarity

Vector databases are built for that.

FAISS vs Vector Database
