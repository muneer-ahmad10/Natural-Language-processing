Suppose you have:

10,000 HR documents
50,000 Finance documents
100,000 Engineering documents

User asks:

"What is the leave policy?"

Without metadata filtering:

FAISS searches 160,000 documents

With metadata filtering:

Department = HR
        ↓
FAISS searches only 10,000 documents


Benefits:

Faster retrieval ⚡
Better relevance 🎯
Lower cost 💰
Better security 🔒
