Suppose you have:

10,000 HR documents
50,000 Finance documents
100,000 Engineering documents

User asks:

"What is the leave policy?"

Without metadata filtering:

FAISS searches 160,000 documents

With metadata filtering:

Department = HR
        ↓
FAISS searches only 10,000 documents


Benefits:

Faster retrieval ⚡
Better relevance 🎯
Lower cost 💰
Better security 🔒

# **Scenario**

```python

!pip install faiss-cpu


import faiss
import numpy as np
```

```python

!pip install sentence-transformers

from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
documents = [
    {
        "text": "Employees get 20 annual leave days.",
        "department": "HR"
    },
    {
        "text": "Quarterly revenue increased by 15%.",
        "department": "Finance"
    },
    {
        "text": "Python is used in backend services.",
        "department": "Engineering"
    }
]
```

```python
#if user asks

question = "What is the leave policy?"

#Faiss would search through
# HR
# Finance
# Engineering..... all together
```

**Practical Filtering**

```python
#Before creating embedding
department='HR'

filtered_docs=[]

for doc in documents:
  if doc['department']== 'HR':
    filtered_docs.append(doc)
```

```python
filtered_docs
```

```python
#Now only HR documents reamin
```

```python
#creating chunks

chunks=[
    doc['text']
    for doc in filtered_docs
]
```

```python
chunks
```

```python
embeddings = embedding_model.encode(chunks)
```

```python
dimension=embeddings.shape[1]
```

```python
dimension
```

```python
index=faiss.IndexFlatL2(dimension)
```

```python
index.add(
    np.array(embeddings).astype("float32")
)
```

```python
index
```

```python
query = "What is the leave policy?"


query_embedding = embedding_model.encode([query])

k = 1
distances, indices = index.search(np.array(query_embedding).astype("float32"), k)

print(f"Question: {query}\n")
print("Most Relevant Match:")
print(chunks[indices[0][0]])
```
